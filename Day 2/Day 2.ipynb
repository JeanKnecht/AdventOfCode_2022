```python
def RPS(input_path):
    with open(input_path, 'r') as ifp:
        l_points = {'X':1, 'Y':2, 'Z':3}
        l_win= {'A':'Y', 'B':'Z', 'C':'X'}
        l_draw = {'A':'X', 'B':'Y', 'C':'Z'}
        l_outcome = {'X':0, 'Y':3, 'Z':6}
        s = [line[:-1].split(" ") for line in ifp]
        points = []
        for i in s:
            if(i[1] == l_win[i[0]]):
                points.append(6+l_points[i[1]])
                continue
            elif(i[1] == l_draw[i[0]]):
                points.append(3+l_points[i[1]])
                continue
            points.append(l_points[i[1]])
            
    return sum(points)
```

```python
RPS("input.txt")
```
